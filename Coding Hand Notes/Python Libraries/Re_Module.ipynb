## Regex
- .       - Any Character Except New Line
- \d      - Digit (0-9)
- \D      - Not a Digit (0-9)
- \w      - Word Character (a-z, A-Z, 0-9, _)
- \W      - Not a Word Character
- \s      - Whitespace (space, tab, newline)
- \S      - Not Whitespace (space, tab, newline)
- \b      - Word Boundary
- \B      - Not a Word Boundary
- ^       - Beginning of a String
- $       - End of a String
- []      - Matches Characters in brackets
- [^ ]    - Matches Characters NOT in brackets
- |       - Either Or
- ( )     - Group

## Quantifiers:
- *       - 0 or More
- +       - 1 or More
- ?       - 0 or One  or  declaring a non-greedy match
- {3}     - Exact Number
- {3,4}   - Range of Numbers (Minimum, Maximum)

- (.*) - anything
- (.*?) - anything in non greedy form
- *? - 0 or more in non greedy form 
- +? - one or more non greedy form



## Repetitions with Braces:
- (Ha){3} - match the string 'HaHaHa'
- (Ha){3,5} - match strings 'HaHaHa' 'HaHaHaHa' 'HaHaHaHaHa'
- (Ha){3,} - match 3 or more 'Ha' Group
- (Ha){,5} - match 0-5 'Ha' Group 
- (Ha){3,5}? - non greedy form i.e. match 3 instances of Ha

- re.compile() second arguments:
- re.DOTALL - make the dot character match all characters, including the newline character
- re.I or re.IGNORECASE - case insensitive matching 
- re.VERBOSE - ignore Whitespace and comments inside the regular expressions string

- To use three options at the same time, use "|"
- re.compile('foo', re.IGNORECASE|re.DOTALL|re.VERBOSE)
